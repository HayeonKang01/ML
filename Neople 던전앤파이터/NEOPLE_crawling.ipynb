```python
title = []
# def neople_crw(link, review_page):
link = url 
review_page = 1000

# webdriver_manager 
dr = webdriver.Chrome(ChromeDriverManager().install())
dr.get('{0}'.format(link))

act = ActionChains(dr)  # 드라이버에 동작을 실행시키는 명령어를 act로 지정

#title = []

start = time.time()
### 예시 ###

print('REVIEW_PAGE : ', review_page)

for i in range(1, review_page): # 리뷰 최대 2000개 까지라고 해서 100p로 제한
    print(i)

    #dr.execute_script("window.scrollTo(0, 10)")
    #time.sleep(10)

    if i == 1 : 
        pass

    elif i % 10 == 1: # 11, 21, 31... 
        #element2 = 
        dr.find_element_by_css_selector('#paging > div > button.-next').send_keys(Keys.ENTER)  # 동작 할 요소 선택
        #act.click(element2).perform() # 위 send_keys가 에러나면 사용하기
        time.sleep(3)

    elif i < 11:
        dr.find_element_by_css_selector('#paging > div > button:nth-child({0})'.format(i+1)).send_keys(Keys.ENTER)
        time.sleep(2)

    else:
        #element1 = 
        dr.find_element_by_css_selector('#paging > div > button:nth-child({0})'.format(i%10+1)).send_keys(Keys.ENTER)  # 동작 할 요소 선택
        #act.click(element1).perform()  # element1  클릭 동작을 수행
        time.sleep(3)



    mydata = dr.find_elements_by_css_selector('#board_list_wrap > ul > li')

    for data in mydata:
        #print(data.text, '\n')
        title.append(data.text)


    time.sleep(3)


    #print(data, '\n')

#df = pd.DataFrame(np.column_stack(list(data)), columns=['data'])
#df.to_csv('{0}.csv'.format(name))
#df.to_csv('test_neo.csv', encoding='utf-8')

end = time.time()
print(title)

print(f"{end - start:.5f} sec=========================")
```
