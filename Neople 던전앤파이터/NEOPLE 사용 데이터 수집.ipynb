# user 6000명 수집 : user_df(기본정보+어드벤쳐/길드정보)

```python
name_list = ['프링', '센치', '노블', '소라', '영혼', '양양', '강릉', '제주', '우주', '귀살', \
 '악마', '대공', '주먹', '치킨', '늑대', '난나', '하연', '진주', '나다', '오징어', \
 '흠냐', '야생', '우성', '행님', '호랑', '어둠', '던전', '다롱', '황제', '졸려']
```

```python
for i, name in enumerate(name_list):
    print(i, name)
```

```python
#userName = '졸려' # 무조건 두글자 이상
user_df = pd.DataFrame()

for i, userName in enumerate(name_list):
    userName = parse.quote(userName)
    url = 'https://api.neople.co.kr/df/servers/all/characters?characterName={0}&limit=200&wordType=full&apikey=OJnvu6JgglMMwk2pvu1jiWQLBTozkx25'.format(userName)

    response = requests.get(url)
    contents = response.text
    json_ob = json.loads(contents)
    user = json_normalize(json_ob['rows'])
    user_df = pd.concat([user_df, user])
```

```python
user_df.reset_index(drop=True, inplace=True)
```

```python
user_df.to_csv('user_df.csv', encoding='utf-8')
```
